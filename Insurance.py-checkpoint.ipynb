```python
import csv
import numpy as np
import pandas as pd
data = pd.read_csv(r'C:\Users\CS\Downloads\insurance_data.csv')
data.head(28)
# data.ndim
# data.dtype
# data.shape
```

```python
# Query-------------------------------------------------------------------------------
import matplotlib.pyplot as plt
print("Creating a plot between age and bought_insurance: ")
r = data.groupby('age')['bought_insurance'].agg(['count'])
dict = r.to_dict()['count']
x = dict.keys()
y = dict.values()
plt.bar(x,y)
plt.show()
```

```python
# Query-----------------------------------------------------------------------------------
plt.figure(figsize=(15, 5))
fig=data['bought_insurance'].value_counts().sort_index()
plt.title('insurance.csv')
plt.xlabel('age')
plt.ylabel('bought_insurance')
plt.plot(fig.index,fig.values)
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(27,2))
x = np.array(data['age'])
y = data['bought_insurance'].value_counts().sort_index()  

plt.xlabel("age")  
plt.ylabel("bought_insurance")  
plt.title("insurance.csv")  

plt.plot(x, y, '-', color='green')

plt.show()
plt.close()
```

```python
plt.figure(figsize=(27,2))
plt.scatter(data['age'], data['bought_insurance'], alpha=0.7)
plt.title('Insurance.csv')
plt.xlabel('Age')
plt.ylabel('bought_insurance')
plt.grid(True)
plt.show()
```
